```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf_vectorizer = TfidfVectorizer(use_idf=True,stop_words = 'english', norm='l2')

train_matrix = tfidf_vectorizer.fit_transform(train_data)
# print(tfidf_vectorizer.vocabulary_)
# print(tfidf_vectorizer.get_feature_names())
print("idf_vector:",tfidf_vectorizer.idf_)
print("***")
print("matrix_shape:",train_matrix.shape)
print("***")
print("feature names:", tfidf_vectorizer.get_feature_names())
# print((tfidf_vectorizer.stop_words_))

    
```

```python

for i in range(len(test_data)):
    
    print("query:", test_actual[i])
    query = test_data[i]
    tfidf_query = tfidf_vectorizer.transform([query])
#     print(tfidf_query.shape)
    query_similarity = cosine_similarity(tfidf_query, train_matrix)
    max_similarity = max(max(query_similarity))
    print('!!cosine similarity score: ', max_similarity)
    index_max = query_similarity.argmax()
    print("index max:",index_max)
    print("match: ", train_actual[index_max])
    print("##")
```
